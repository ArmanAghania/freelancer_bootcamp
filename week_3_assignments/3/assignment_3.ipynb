```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('e-commerce.csv')
```

```python
df.head()
```

```python
monthly_purchase = df.groupby(['customerNumber', 'orderMonth']).agg({'quantityOrdered': 'sum', 'final_payment': 'sum'}).reset_index()
monthly_purchase
```

```python
customer_number = 496
```

```python
monthly_purchase[monthly_purchase['customerNumber'] == 103]
```

```python
monthly_purchase[monthly_purchase['customerNumber'] == customer_number]
```

```python
fig , ax = plt.subplots(figsize=(10,6))
ax.plot(monthly_purchase[monthly_purchase['customerNumber'] == customer_number ]['orderMonth'], monthly_purchase[monthly_purchase['customerNumber'] == customer_number]['final_payment'], color='#005f73', marker='o')
# ax.fill_between(monthly_purchase[monthly_purchase['customerNumber'] == f'{customer_number}']['orderMonth'], 0, monthly_purchase[monthly_purchase['customerNumber'] == f'{customer_number}']['quantityOrdered'], color='#005f73', alpha=0.6)
ax.set_facecolor('#EAF4f4')
fig.set_facecolor('#EAF4f4')
ax.set_title('Monthly Quantity Ordered per Product')
ax.set_xlabel('Month')
ax.set_ylabel('Quantity Ordered')
ax.spines[['left', 'top']].set_visible(False)
ax.tick_params(length=0)
ax.yaxis.set_ticks_position('right')
plt.legend(loc='upper right', facecolor='#EAF4f4')
plt.show()
```

```python
monthly_customer_purchase = df.groupby(['customerNumber', 'productName']).agg({'quantityOrdered': 'sum'}).reset_index()
monthly_customer_purchase
```

```python
fig , ax = plt.subplots(figsize=(10,6))
ax.plot(monthly_customer_purchase[monthly_customer_purchase['customerNumber'] == customer_number ]['productName'], monthly_customer_purchase[monthly_customer_purchase['customerNumber'] == customer_number]['quantityOrdered'], color='#005f73', marker='o')
ax.set_facecolor('#EAF4f4')
fig.set_facecolor('#EAF4f4')
ax.set_title('Monthly Price Average')
ax.set_xlabel('Month')
ax.set_ylabel('Average Price ($)')
ax.spines[['left', 'top']].set_visible(False)
ax.tick_params(length=0)
ax.yaxis.set_ticks_position('right')
ax.tick_params(axis='x', color='#22333b', rotation=90)
plt.legend(loc='upper right', facecolor='#EAF4f4')
plt.show()
```
