```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('e-commerce.csv')
```

```python
df.head()
```

```python
monthly_count = df.groupby(['productName', 'orderMonth']).agg({'quantityOrdered': 'sum', 'priceEach': 'mean'}).reset_index()
monthly_count
```

```python
product_name = 'The USS Constitution Ship'
```

```python
fig , ax = plt.subplots(figsize=(10,6))
ax.plot(monthly_count[monthly_count['productName'] == f'{product_name}']['orderMonth'], monthly_count[monthly_count['productName'] == f'{product_name}']['quantityOrdered'], color='#005f73', marker='o')
ax.fill_between(monthly_count[monthly_count['productName'] == f'{product_name}']['orderMonth'], 0, monthly_count[monthly_count['productName'] == f'{product_name}']['quantityOrdered'], color='#005f73', alpha=0.6)
ax.set_facecolor('#EAF4f4')
fig.set_facecolor('#EAF4f4')
ax.set_title('Monthly Quantity Ordered per Product')
ax.set_xlabel('Month')
ax.set_ylabel('Quantity Ordered')
ax.spines[['left', 'top']].set_visible(False)
ax.tick_params(length=0)
ax.yaxis.set_ticks_position('right')
plt.legend(loc='upper right', facecolor='#EAF4f4')
plt.show()
```

```python
fig , ax = plt.subplots(figsize=(10,6))
ax.plot(monthly_count[monthly_count['productName'] == f'{product_name}']['orderMonth'], monthly_count[monthly_count['productName'] == f'{product_name}']['priceEach'], color='#005f73', marker='o')
ax.set_facecolor('#EAF4f4')
fig.set_facecolor('#EAF4f4')
ax.set_title('Monthly Price Average')
ax.set_xlabel('Month')
ax.set_ylabel('Average Price ($)')
ax.spines[['left', 'top']].set_visible(False)
ax.tick_params(length=0)
ax.yaxis.set_ticks_position('right')
plt.grid(color='#EAF4f4', linestyle='-', linewidth=1)
plt.legend(loc='upper right', facecolor='#EAF4f4')
plt.show()
```
